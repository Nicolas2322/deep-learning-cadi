```python
import numpy as np
import pandas as pd

# ----------------------------------------
# Definición de pesos y bias
# ----------------------------------------

w1 = 0.6
w2 = 0.4
bias = -0.5

# ----------------------------------------
# Casos de prueba (entradas del modelo)
# ----------------------------------------

pruebas = [
    (0,0),
    (0,1),
    (1,0),
    (1,1)
]

# ----------------------------------------
# Función del perceptrón
# ----------------------------------------

def perceptron(x1, x2, w1, w2, bias):

    # cálculo del puntaje z
    z = (x1 * w1) + (x2 * w2) + bias

    # función de activación (umbral)
    if z >= 0:
        salida = 1
    else:
        salida = 0

    return z, salida


# ----------------------------------------
# Ejecución de pruebas
# ----------------------------------------

resultados = []

for x1, x2 in pruebas:

    z, salida = perceptron(x1, x2, w1, w2, bias)

    resultados.append({
        "Entrada x1": x1,
        "Entrada x2": x2,
        "Valor z": round(z,2),
        "Salida": salida
    })

# ----------------------------------------
# resultados en tabla
# ----------------------------------------

df = pd.DataFrame(resultados)

print("Resultados del Perceptrón\n")

df
```

## Análisis de resultados

Se observa que la salida del perceptrón depende directamente
de los valores de los pesos y del bias.

El bias desplaza el umbral de decisión, mientras que los pesos
determinan la importancia de cada entrada.

Cuando el valor de z es mayor o igual a cero, la neurona
activa la salida (1). Cuando es menor a cero, produce
una salida de 0.

Modificar los pesos o el bias cambia el comportamiento
de clasificación del modelo.

## Conclusiones

El comportamiento del perceptrón depende directamente de los pesos
y del bias utilizados.

Al modificar estos parámetros cambia el valor de z y por lo tanto
la decisión final del modelo.

El perceptrón clasifica las entradas en 0 o 1 utilizando una función
de activación basada en un umbral.
